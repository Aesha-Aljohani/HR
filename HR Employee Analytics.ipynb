```python
sizes = list(df['left'].value_counts().sort_values().values)
labels = list(df['left'].value_counts().sort_values().index)

explode = [0, 0.1]

plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, explode = explode,
        autopct=lambda p:f'{p:.2f}%')
plt.title('The Number of the Employee whose Leave or Stay in there Jobs', fontsize=25)
plt.tight_layout()
plt.show()
```
